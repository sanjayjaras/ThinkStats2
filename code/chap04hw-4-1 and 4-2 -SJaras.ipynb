How much did you weigh at birth? If you donâ€™t know, call your

### Define Functions

```python

def percentileRank(scores, yourScore):
    """ Compute percentile rank
        Retrun:
        float rank"""
    # initialize the count variable
    count = 0
    # find out how many scores lower than or equal to your scores
    for score in scores:
        if score <= yourScore:
            count += 1

    # convert count to percentile rank by dividing total number of scores
    percentile_rank = 100.0 * count / len(scores)
    return percentile_rank
```

### Compute percentile rank

```python
# Filter all non-living children
liveChild = preg[preg.outcome == 1]
# filter dataset for non-first children
otherChilds = liveChild[liveChild.birthord != 1]
weightsNonFirstChilds = otherChilds["totalwgt_lb"]
# My Weight
myWeight = 7.6
percentileRankForMe = percentileRank(weightsNonFirstChilds, myWeight)
print("My Percentile Rank", percentileRankForMe)
```

### Exercise 4.2 
The numbers generated by random.random are supposed to be
uniform between 0 and 1; that is, every value in the range should have the
same probability.
Generate 1000 numbers from random.random and plot their PMF and CDF.
Is the distribution uniform?

### Generate random numbers

```python
# inialize list for random numbers
randomNumbers = []
# generate random number and add it to list
for i in range(1, 1001):
    randomNumbers.append(random.random())
```

### Draw CDF for random numbers

```python
cdfRandom = thinkstats2.Cdf(randomNumbers)
thinkplot.Cdf(cdfRandom)
thinkplot.Show(xlabel='Random Numbers', ylabel='CDF')
```

### Draw PMF for random numbers

```python
pmfRandom = thinkstats2.Pmf(randomNumbers)
thinkplot.Pmfs([pmfRandom])
thinkplot.Show(xlabel='Random Numbers', ylabel='PMF')
```

### Conclusion
The distribution is perfectly normal as shown by both plots
PMF shows each number range has 1% distribution same is shown by CDF plot.
